### Cleaning up data

```python
import pandas as pd

df = pd.read_csv ('Womens_Clothing_E-Commerce_Reviews.csv')
print(df)
```

```python
df.fillna("untitled")
```

```python
# Create a Ratings list.
Rating = df["Rating"].tolist()
Rating
```

```python
# Create bins in which to place values based Ratings
bins = [0, 1, 2, 3, 4, 5, 6]

# Create labels for these bins
group_labels = ["Negative", "Negative", "Negative", "Positive", "Positive", "Positive"]
```

```python
# Slice the data and place it into bins
pd.cut(df["Rating"], bins, labels=group_labels, ordered=False).head()
```

```python
# Place the data series into a new column inside of the DataFrame
df["Positive or Negative"] = pd.cut(df["Rating"], bins, labels=group_labels, ordered=False)
df.head()
```

```python
df2 = df[['Positive or Negative', 'Review Text']].copy()
```
